#### 4. Subplots: Two options for subplots

```python
# Subplot option 1
fig, ( (ax1, ax2), (ax3, ax4)) = plt.subplots( nrows = 2,
                                              ncols =2,
                                              figsize = (10, 5)
                                             )
# Plot data to each axis
ax1.plot(x,x/2)
ax1.set(title="Simple plot");
ax2.scatter(np.random.random(10), np.random.random(10))
ax2.set(title="Scatter plot");
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.set(title="Bar plot");
ax4.hist(np.random.randn(1000))
ax4.set(title="Histogram");
```

```python
# Subplots option 2
fig, ax= plt.subplots( nrows = 2,
                       ncols =2,
                       figsize = (10, 5)
                    )

# plot data to each axis
ax[0,0].plot(x,x/2)
ax[0,0].set(title="Simple plot");
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[0,1].set(title="Scatter plot");
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,0].set(title="Bar plot");
ax[1,1].hist(np.random.randn(1000))
ax[1,1].set(title="Histogram");
```

### Which one should I use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, it is okay to use the pyplot method.
* When plotting something more advanced, use the OO method.


```python
# Patients with age over 50
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

#### Scatter plot

```python
# Pyplot method
over_50.plot(x="age",
             y="chol",
             c="target",
             kind="scatter"
            );
```

```python
# Object Oriented method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(x="age",
             y="chol",
             c="target",
             kind="scatter",
             ax=ax
            )
ax.set(xlim=[48, 80])
```

```python
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = "--");
```

#### Subplots

```python
fig, (ax0, ax1) = plt.subplots(nrows =2,
                               ncols =1,
                               figsize=(10,10),
                              sharex=True)

#Add data to ax0 
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle = "--");



#Add data to ax1 
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);
# customize ax1
ax1.set(title = "Heart Disease and Maximum Heart Rate",
       xlabel="Age",
       ylabel="thalach")

# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# Add a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter");  # This changes the colour scheme
# customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = "--");
```

This plot shows some information about the heart disease dataset...

```python
# Customizing the y and x axses limits

fig, (ax0, ax1) = plt.subplots(nrows =2,
                               ncols =1,
                               figsize=(10,10),
                              sharex=True)

#Add data to ax0 
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");
# customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# change the x-axis limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle = "--");



#Add data to ax1 
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter");
# customize ax1
ax1.set(title = "Heart Disease and Maximum Heart Rate",
       xlabel="Age",
       ylabel="thalach")

# change the x-axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# Add a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
